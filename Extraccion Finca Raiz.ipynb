```python
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains
import math
import time
import concurrent.futures
import pandas as pd
```

```python
habitaciones=[]
banios=[]
area_const=[]
area_priv=[]
estrato=[]
antiguedad=[]
piso=[]
parqueaderos=[]
admin=[]
precio_m2=[]
precio=[]
direccion=[]
barrio=[]
```

```python
def pagina(url):
    options =  webdriver.ChromeOptions()
    options.add_argument('--start-maximized')
    options.add_argument('--disable-extensions')

    ser = Service(r'D:\Descargas\chromedriver_win32 (1)\\chromedriver.exe')

    driver = webdriver.Chrome(service=ser, options=options)

    driver.get(url)
    
    padre = driver.window_handles[0]
        
    for i in range(32):
        
        #/html/body/div[1]/div[1]/div[3]/div/div[1]/div[2]/div[1]
        
        #/html/body/div[1]/div[1]/div[3]/div/div[1]/div[3]
        
        try:
            WebDriverWait(driver, 10)\
                    .until(EC.element_to_be_clickable((By.XPATH,
                                                      f"/html/body/div[1]/div[1]/div[3]/div/div/div[2]/div[2]/div[{i+1}]")))\
                    .click()
            time.sleep(0.5)

            hijo = driver.window_handles[1]

            driver.switch_to.window(hijo)                  

            try:

                grid = driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/main/section/div[4]/div/section/div[1]/div/div[2]/div[1]")
                list_grid = grid.text.split('\n')

                llaves = list_grid[0::2]
                valores = list_grid[1::2]
                diccionario = dict(zip(llaves, valores))

                if "Habitaciones" not in diccionario:
                    habitaciones.append(0) 
                else:
                    habitaciones.append(diccionario["Habitaciones"])

                if "Baños" not in diccionario:
                    banios.append(0) 
                    
                else:
                    banios.append(diccionario["Baños"])

                if "Área construída" not in diccionario:
                    area_const.append('Sin informacón')  
                else:
                    area_const.append(diccionario["Área construída"])

                if "Área privada" not in diccionario:
                    area_priv.append('Sin informacón')  
                else:
                    area_priv.append(diccionario["Área privada"])

                if "Estrato" not in diccionario:
                    estrato.append(0) 
                else:
                    estrato.append(diccionario["Estrato"])

                if "Antigüedad" not in diccionario:
                    antiguedad.append('Sin informacón')  
                else:
                    antiguedad.append(diccionario["Antigüedad"])

                if "Piso N°" not in diccionario:
                    piso.append(1) 
                else:
                    piso.append(diccionario["Piso N°"])

                if "Administración" not in diccionario:
                    admin.append('Sin informacón') 
                else:
                    admin.append(diccionario["Administración"])

                if "Precio m²" not in diccionario:
                    precio_m2.append(0) 
                else:
                    precio_m2.append(diccionario["Precio m²"])
                    
                if "Parqueaderos" not in diccionario:
                    parqueaderos.append(0) 
                else:
                    parqueaderos.append(diccionario["Parqueaderos"])

                try:
                    barrio_sc = driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/main/section/div[4]/div/section/div[3]/div/div[2]")

                    list_barrio = barrio_sc.text.split('\n')
                    
                    llaves_1 = list_barrio [0::2]
                    valores_1 = list_barrio [1::2]
                    diccionario_1 = dict(zip(llaves_1, valores_1))


                    if "Dirección" not in diccionario_1:
                        direccion.append('Sin información') 
                    else:
                        direccion.append(diccionario_1["Dirección"])

                    if "Barrio" not in diccionario_1:
                        barrio.append('Sin información') 
                    else:
                        barrio.append(diccionario_1["Barrio"])

                except:
                    direccion.append('Sin información')
                    barrio.append('Sin información')
                    
                precio_sc = driver.find_element(By.XPATH,"/html/body/div[1]/div[2]/main/section/div[4]/div/aside/div/div[1]/div[1]")

                try:
                    llaves_2 = precio_sc.text.split('\n') [0::2]
                    valores_2 = precio_sc.text.split('\n') [1::2]
                    diccionario_2 = dict(zip(llaves_2, valores_2))

                    if "Precio total (COP)" not in diccionario_2:
                        precio.append('Sin información') 
                    else:
                        precio.append(diccionario_2["Precio total (COP)"])

                except:
                    precio.append('Sin información') 
                                        
                driver.close()

                driver.switch_to.window(padre)

            except:

                driver.close()

                driver.switch_to.window(padre)


        except:
                       
            print(f'error en la pesataña {i+1}')
            
        
        
        df = pd.DataFrame({'habitaciones': habitaciones, 
                   'banios': banios, 
                   'area_const':area_const,
                   'area_priv':area_priv,
                   'estrato':estrato,
                   'antiguedad':antiguedad,
                   'piso':piso,
                   'admin':admin,
                   'precio_m2':precio_m2,
                   'parqueaderos':parqueaderos,
                   'precio':precio,
                   'direccion':direccion,
                   'barrio':barrio})
        
    driver.quit()
        
#         return df
```

```python
urls = [f'https://www.fincaraiz.com.co/apartamentos/venta/bogota/bogota-dc?pagina={i+1}' for i in range (0,5,1)]
```

```python
with concurrent.futures.ThreadPoolExecutor(max_workers=5) as executor:

    executor.map(pagina, urls)
```
